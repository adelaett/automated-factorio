```python
import sklearn
import sklearn.model_selection
import sklearn.linear_model
import sklearn.datasets
import numpy as np

class MySGD(sklearn.linear_model.SGDRegressor):
    pass
    # This estimator defines in particular the coef_ and t_ attributes once fitted.

def score_mean_coef(estimator, X, y):
    return np.mean(estimator.coef_)

def score_time_to_fit(estimator, X, y):
    return np.min(estimator.t_)


X, y = sklearn.datasets.load_digits(return_X_y=True)

model = sklearn.model_selection.GridSearchCV(
    MySGD(),
    {
        "alpha": [0.0001, 0.01, 0.1, 1, 10]
    },
    scoring={
        "mean_coef": score_mean_coef,
        "steps_to_converge": score_time_to_fit
    },
    refit=False
).fit(X, y)

import pandas as pd

pd.DataFrame({k: v for k, v in model.cv_results_.items() if not k.startswith("split")})
```
